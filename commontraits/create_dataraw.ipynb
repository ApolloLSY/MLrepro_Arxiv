```python
df = pd.read_csv("/data/Clockbase/Methylation/sample_disease.txt",sep=" ", header=0)  
result_df = pd.DataFrame()
#filter samples with disease label
selected_columns = df.columns[df.iloc[0].notna()]
result_df = pd.concat([result_df, df[selected_columns]], axis=1)

result_df.head()
```

```python
#prepare a dataframe
data = pd.DataFrame(index=result_df['sample_id'])
data['disease'] = result_df['disease']
data['age'] = result_df['age']
data['sex'] = result_df['sex']
display(data)
```

```python

# get CpGs data
# set path
folder_path = "/data/Clockbase/Methylation/disease/"

# find path
file_names = os.listdir(folder_path)

# find GSE files
for file_name in file_names:
    if file_name.startswith("GSE"):  
        file_path = os.path.join(folder_path, file_name)
        # load CpGs data
        dfCpGs = pd.read_csv(file_path,compression='gzip')   
        dfCpGs_trans=dfCpGs.transpose()
        # map CpGs to profile
        data = data.join(dfCpGs_trans.iloc[:, 2:], how='left')

       
```

```python
output_path="/data/Clockbase/Methylation/disease_predictor/"
if not os.path.exists(output_path):
    os.makedirs(output_path)
data.to_csv(os.path.join(output_path, 'dataraw.csv'),header=0, float_format='%.4f', index=True)
```
